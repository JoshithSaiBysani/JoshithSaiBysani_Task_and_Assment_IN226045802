## User Login Check
#### Problem Statement:
Given a username and password, check whether login is successful.
- username = "admin"
- password = "1234"
#### Requirements:
- Print "Login Successful" if both username and password match
- Otherwise print "Invalid Credentials

```python
username = input("Enter the username")
password = input("Enter the password")
if(username == "admin" and password == "1234"):
    print("Login Successful")
else:
    print("Invalid Credentials")
```

## Pass / Fail Analyzer
#### Problem Statement:
- Analyze student results.
- marks = [45, 78, 90, 33, 60]

#### Requirements:
- A student passes if marks â‰¥ 50
- Count the total number of pass students
- Count the total number of fail students
- Print the final result clearly

```python
marks = [45, 78, 90, 33, 60]
pass_count = fail_count = 0
for mark in marks:
    if mark >= 50:
        pass_count += 1
    else:
        fail_count += 1
print("Number of students passed: ",pass_count)
print("Number of students failed: ",fail_count)
```

## Simple Data Cleaner
#### Problem Statement:
- Clean and standardize user names.
- names = [" Alice ", "bob", " CHARLIE "]

#### Requirements:
- Remove extra spaces
- Convert all names to lowercase
- Print the cleaned list

```python
names = [" Alice ", "bob", " CHARLIE "]
names_cleaned = []
for name in names:
    names_cleaned.append(name.strip().lower()) 
    # strip() is used to remove leading and trailing spaces from a string
    # lower() converts the given string to lowercase
    # append() function used to add values at the end of list
print(names_cleaned)
```

## Message Length Analyzer
#### Problem Statement:
- Analyze message sizes.
- messages = ["Hi", "Welcome to the platform", "OK"]

#### Requirements:
- Print the length of each message
- Flag messages longer than 10 characters

```python
messages = ["Hi", "Welcome to the platform", "OK"]
flag = False
for message in messages:
    print(message, "- Length =", len(message))
    if(len(message) > 10):
        flag = True
        print(message, ": Flagged, message too long")
```

## Error Message Detector
#### Problem Statement:
- Detect error messages from system logs.
- logs = ["INFO", "ERROR", "WARNING", "ERROR"]

#### Requirements:
- Count the number of "ERROR" entries
- Print the total error count

```python
logs = ["INFO", "ERROR", "WARNING", "ERROR"]
error_count = 0
for log in logs:
    if(log == "ERROR"):
        error_count += 1
print("Total error count =", error_count)
```
