### Problem 1: Social Media – Post Engagement Analyzer
#### Problem Statement
- Create a function to analyze engagement on a social media post.
#### Rules
- Calculate total likes
- If Total Likes ≥ 1000 → Viral Post
- Otherwise → Normal Engagement
#### Expected Output Format
- Total Likes: 1050
- Post Status: Viral Post

```python
def analyze_engagement(likes): # Function definition
    total = 0
    for i in likes:
        total += i
    if total >= 1000:
        print("Post Status: Viral Post")
    else:
        print("Post Status: Normal Engagement")
n = int(input("Enter number of days for analysis"))
likes = list(map(int, input("Enter the number of likes").split()))
analyze_engagement(likes) # Function call
```

### Problem 3: Agriculture – Rainfall Adequacy Checker
#### Problem Statement
- Create a function that analyzes seasonal rainfall data.
#### Rules
- Calculate average rainfall
- Average ≥ required level → Adequate Rainfall
- Otherwise → Inadequate Rainfall
#### Expected Output Format
- Average Rainfall: 72
- Rainfall Status: Adequate Rainfall

```python
def analyze_rainfall(rain):
    threshold = 70
    total = 0
    for i in rain:
        total += i
    avg_rainfall = total / len(rain)
    if avg_rainfall > threshold:
        return "Adequate Rainfall"
    else:
        return "Inadequate Rainfall"
n = int(input("Enter number of days for analysis"))
rainfall = list(map(int, input("Enter the rainfall amount").split()))
print("Rainfall Status:",analyze_rainfall(rainfall)) 
```

### Problem 4: Social Media – Duplicate Account Detection
#### Problem Statement
- Create a function to detect duplicate usernames in a system.
#### Rules
- If duplicates exist → Flag accounts
- Else → Accounts are unique
#### Expected Output Format
- Duplicate Accounts Found: Yes

```python
def detect_duplicates(accounts):
    unique = set(accounts)
    if len(accounts) == len(unique):
        print("Duplicate Accounts Found: No")
    else:
        print("Duplicate Accounts Found: Yes")
usernames = list(map(str, input("Enter usernames").lower().split()))
detect_duplicates(usernames)
```

### Problem 6: Agriculture – Premium Crop Price Filter
#### Problem Statement
- Create a function that filters crop prices above a premium threshold.
#### Rules
- Price > ₹2000 → Premium Crop
- Else → Regular Crop
#### Expected Output Format
- Premium Crops: [2500, 3200]

```python
def filter_crops(crops):
    premium_crops = []
    for price in crops:
        if price > 2000:
           premium_crops.append(price)
    return premium_crops
n = int(input("Enter the number of crops"))
crops = list(map(int, input("Enter price of each crop").split()))
print("Premium Crops:",filter_crops(crops))
```

### Problem 7: System Monitoring – Application Health Checker
#### Problem Statement
- Create a function to evaluate application health based on error count.
#### Rules
- Errors = 0 → Healthy
- Errors ≤ 5 → Minor Issues
- Errors > 5 → Critical Issues
#### Expected Output Format
- Error Count: 7
- System Status: Critical Issues

```python
def evaluate_health(errors):
    if errors == 0:
        return "Healthy"
    if errors <= 5:
        return "Minor Issues"
    else:
        return "Critical Issues"
errors = int(input("Enter the errors"))
print()
print("Error Count:", errors)
print("System Status:", evaluate_health(errors))
```
